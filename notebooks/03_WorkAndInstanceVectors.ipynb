# Create Triple Embeddings for BIBFRAME Works and Instances

```python
import httpx

from tqdm.notebook import tqdm
from helpers import get_token
```

```python
works_page_01 = httpx.get("http://localhost:8000/api/change_documents/works/page/1")
```

```python
work_errors = []
for work_row in tqdm(works_page_01.json()['orderedItems']):
    work_url = work_row['object']['id']
    embeddings_url = f"{work_url}/embeddings"
    embeddings_result = httpx.post(embeddings_url,
                                   timeout=120,
                                   headers={
                                    "Authorization": f"Bearer {get_token()}"
                                    })
    try:
        embeddings_result.raise_for_status()
    except:
        work_errors.append(work_url)
```

```python
len(work_errors)
```

```python
instances_page_01 = httpx.get("http://localhost:8000/api/change_documents/instances/page/1")
```

```python
instance_errors = []
for instance_row in tqdm(instances_page_01.json()['orderedItems']):
    instance_url = instance_row['object']['id']
    embeddings_url = f"{instance_url}/embeddings"
    embeddings_result = httpx.post(embeddings_url,
                                   timeout=120,
                                   headers={
                                    "Authorization": f"Bearer {get_token()}"
                                    })
    try:
        embeddings_result.raise_for_status()
    except:
        instance_errors.append(work_url)
```
