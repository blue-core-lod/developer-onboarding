# Load Sample of LOC CBD files

```python
import zipfile

import httpx

from helpers import get_token
from tqdm.notebook import tqdm
```

```python
with zipfile.ZipFile("loc-sample.zip") as loc_cbd_zip:
    cbd_files = loc_cbd_zip.namelist()
    errors = []
    for name in tqdm(cbd_files):
        with loc_cbd_zip.open(name) as cbd_file:
            files = {"file": cbd_file.read()}
            batch_post = httpx.post("http://localhost:8000/api/batches/upload/",
                                headers={
                                   "Authorization": f"Bearer {get_token()}"
                                },
                                follow_redirects=True,
                                files=files)
            try:
                batch_post.raise_for_status()
            except:
                errors.append(name)
    print(f"\nFinished loading {len(cbd_files)}, total errors: {len(errors)}")
```
